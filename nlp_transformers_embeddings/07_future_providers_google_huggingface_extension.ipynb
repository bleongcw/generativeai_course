## Concept primer

Provider portability does not mean every provider has the same features. A clean course architecture names the capability needed: embeddings, generation, evaluation, tool use, or reranking. Then each provider adapter implements the capabilities it actually supports.

## Step 2 - Define capability-oriented adapter stubs

These are intentionally small. They are not full implementations yet; they prepare the course for future modules without forcing today’s notebooks to change.


```python
from dataclasses import dataclass
from typing import Protocol

class EmbeddingProvider(Protocol):
    name: str
    def embed(self, texts: list[str]) -> list[list[float]]:
        ...

class GenerationProvider(Protocol):
    name: str
    def generate(self, prompt: str) -> str:
        ...

@dataclass
class FutureGoogleEmbeddingProvider:
    name: str = "google-ai-embedding-future"
    def embed(self, texts: list[str]) -> list[list[float]]:
        # Future implementation: call Google embedding API with secure env-based credentials.
        raise NotImplementedError("Add Google AI embedding implementation in a future module.")

@dataclass
class FutureHuggingFaceEmbeddingProvider:
    name: str = "huggingface-local-embedding-future"
    def embed(self, texts: list[str]) -> list[list[float]]:
        # Future implementation: load a sentence-transformers model or call Hugging Face Inference API.
        raise NotImplementedError("Add Hugging Face embedding implementation in a future module.")

@dataclass
class OptionalVoyageEmbeddingProvider:
    name: str = "voyage-embedding-optional"
    def embed(self, texts: list[str]) -> list[list[float]]:
        # Anthropic documentation points to Voyage AI as an embedding option.
        raise NotImplementedError("Add Voyage AI only if the class needs an Anthropic-adjacent embedding path.")

print("Adapter stubs ready for future modules.")
```

## Debugging checklist

- Do not add provider-specific code inside lesson logic; put it in adapters.
- Do not assume a provider supports embeddings just because it supports chat.
- Do not require optional future keys for today’s notebooks.

## Production best practices

- Design around capabilities, not brand names.
- Keep credentials provider-specific and loaded from `.env`.
- Store model names and provider versions with outputs.
- Add conformance tests for every new adapter.
